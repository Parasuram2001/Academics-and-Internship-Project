```python
import pandas as pd
path="C:\\Users\\parashuram\\Desktop\\ipro\\heart.csv"
data=pd.read_csv(path)

import sklearn
from sklearn.preprocessing import LabelEncoder
le_Sex=LabelEncoder()
le_ChestPainType=LabelEncoder()
le_RestingECG=LabelEncoder()
le_ExerciseAngina=LabelEncoder()
le_ST_Slope=LabelEncoder()
data["Sex_n"]=le_Sex.fit_transform(data["Sex"])
data["ChestPainType_n"]=le_ChestPainType.fit_transform(data["ChestPainType"])
data["RestingECG_n"]=le_RestingECG.fit_transform(data["RestingECG"])
data["ExerciseAngina_n"]=le_ExerciseAngina.fit_transform(data["ExerciseAngina"])
data["ST_Slope_n"]=le_ST_Slope.fit_transform(data["ST_Slope"])

print(data)

inputs=data.drop(["HeartDisease","Sex","ChestPainType","RestingECG","ExerciseAngina","ST_Slope"],"columns")
print(inputs)
outputs=data.drop(["Age","Sex","ChestPainType","RestingBP","Cholesterol","FastingBS","RestingECG","MaxHR","ExerciseAngina","Oldpeak","ST_Slope","Sex_n","ChestPainType_n","RestingECG_n","ExerciseAngina_n","ST_Slope_n"],"columns")
print("outputs",outputs)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,train_size=0.8)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
 
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print("y_pred",y_pred)
print("y_test",y_test)

res=model.predict([[48,138,214,0,108,1.5,0,0,1,1,1]])
#res=model.predict([[int(Age),int(Restingbp),int(Cholesterol),int(Fastingbs),int(Maxhr),float(Oldpeak),int(Sex),int(Chestpaintype),int(Restingecg),int(Exerciseangina),int(STslope)]])
print("result",res)
```
